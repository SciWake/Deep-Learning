<h1 style="color:#008B8B">1. Полносвязный слой</h1>

На входе $n$ чисел, на выходе слоя $m$ чисел:

* $\large x_1, \ldots, x_n$ - входы

* $\large z_1, \ldots, z_m$ - выходы

Как устроен слой, который выполняет данное преобразование? Каждый выход вычисляется как линейная модель над входами:

$\large z_j = \sum\limits_{i=1}^n w_{ji} x_i + b_j$, где $\large w_{ji}, b_j$ - Это некоторые параметры, которые подбираются под данные. Нейрон - это одна модель в составе полносвязного слоя, значит, данная модель является нейроном (на картинке обозначены кругом).

Полносвязный слой имеет много параметров, так как таких моделей будет $m$ штук, а параметров $n + 1$, получается всего параметров будет $m(n + 1)$:

* $m$ линейных моделей, в каждой $(n + 1)$ параметров

* Всего примерно $mn$ параметров в полносвязном слое

* Это очень много: если у нас 1.000.000 входных признаков и 1000 выходов, то это 1.000.000.000 параметров, значит надо много данных

КАРТИНКА

**Почему слой называется полносвязным?**

Видим, что все выходные элементы (нейроны) слоя связаны с элементами входного слоя ($z_1$ зависит от $x_1, x_2, x_3$, аналогично). На данной картинке два полносвзяных слоя, первый слой это преобразование входов в выходы $z_m$ и так далее. 

**Как объеденить слои в одну модель?**

Рассмотрим два полносвязных слоя, первый считаем как $z$, а второй как $s$. У нас есть выход второго полносвязного слоя $s_k$. Подставим в $z_j$ результат предыдущего полносвязного слоя (формулу выше):

$\large s_k = \sum\limits_{j=1}^m v_{kj} z_j + c_k = \sum\limits_{j=1}^m v_{kj} \sum\limits_{i=1}^n w_{ji} x_i + \sum\limits_{j=1}^m v_{kj} b_j + c_k =$

Выполняем преобразования и получаем, что выход второго полносвязного $s_k$ слоя это линейная комбинация входов нейронной сети с некоторыми коэффициентами:

$\large = \sum\limits_{j=1}^m(\sum\limits_{i=1}^n v_{kj}w_{ji}x_i + v_{kj}b_j + \frac{1}{m}c_k)$

Это означает, что любое количество полносвязных слоев не усложняет модель, она как прежде отстанется линейной моделью с некоторыми коэффициентами (Композиция полносвязных слоев - это один полносвязный слой).

**Как усложнить модель?**

Нужно добавлять нелинейную функцию после полносвязного
слоя:

$\large z_j = f(\sum\limits_{i=1}^n w_{ji} x_i + b_j)$

* Вариант 1: $\large f(x) = \frac{1}{1 + \exp(-x)}$ (Сигмоида)

* Вариант 2: $\large f(x) = \max(0, x)$ (ReLU, REctified Linear Unit)
