<h1 style="color:#008B8B">1 Receptive field</h1>

Имеется сверточный слой, который на вход получает картинку и на выход выдает картинку. Берем некоторый пиксель на выходной картинке и задаемся вопросм, от скольки пикселей входной картинки зависит данный выходной пиксель.

* Возьмём пиксель в итоговом изображении (после свёрточных слоёв)
* От какой части входного изображения зависит значение в этом пикселе?

<h2 style="color:#008B8B">1.2 Поле восприятия для свертки $3 \times 3$</h2>

### Один сверточный слой

Входную картинку мы сварачиваем с использованием фильтра 3 на 3 и получаем некоторую выходную картинку. Берем некоторый выходной пиксель и смотрим, от каких пикселей входного изображения он зависит. Чтобы получить данный пиксель, мы с центром входного пикселя (красный пиксель) поставили фильтр (зеленая область), после, этот входной пиксель с текущим фильтром мы свернули в выходной. Значит, у этого пикселя поле восприятия $3 \times 3$.

<img src='img/lecture03/1.png'>

### Два сверточных слоя

Левую картинку мы сварачиваем с использованием фильтра размером $3 \times 3$ и получаем картинку, которую мы сварачиваем с использованием другого фильтра размером $3 \times 3$.

<img src='img/lecture03/2.png'>

Какое поле восприятия у выходного красного пикселя? Красный пиксель мы получили на области из 9 пикселей промежуточного изображения (свернутого первым фильтром). 

А как мы получили каждый из этих 9 пикселей? Мы можем взять первый выходной пиксель, который обозначен фиолетовым цветом и посмотреть от каких входных пикселей он зависит: для этого берем входной пиксель с этими же координатами, относительно его центра на входном изображении поставить фильтр и свернем его в выходной фиоллетовый пиксель. Дальше берем каждый выходной пиксель первого сверточного слоя (зеленая область) и смотрим, от каких пикселей входного изображения он зависит и в совокупности мы получим синюю область.

**Поле восприятия для свёртки 3 x 3:**

* После 1 свёрточного слоя: 3 x 3
* После 2 свёрточных слоев: 5 x 5
* После 3 свёрточных слоёв: 7 x 7

Мы можем увеличивать размер поля восприятия за счет увеличения количества сверточных слоев подряд. Но размер поля восприятия возврастает довольно медленно с увеличением количества сверточных слоев подряд. Для этого можно вязть свертки большего размера:

<h2 style="color:#008B8B">1.2 Поле восприятия для свертки $5 \times 5$</h2>

### Два сверточных слоя

Левую картинку мы сварачиваем с использованием фильтра размером $5 \times 5$ и полученную картинку мы сварачиваем с использованием другого фильтра размером $5 \times 5$.

<img src='img/lecture03/3.png'>

Тогда выходной красный пиксель зависит от входной зеленой области размером $5 \times 5$. И каждый пиксель зеленой обласит является выходным для первого сверточного слоя. Остается определить то, от какой входной области каждый выходной пиксель зависит и в совокупности мы получаем синюю область. А синяя область определяет зону восприятия карсного пикселя выходного второго сверточного слоя.

**Поле восприятия для свёртки 5 x 5:**

Поле восприятия для свёртки 5 x 5:
* После 1 свёрточного слоя: 5 x 5
* После 2 свёрточных слоев: 9 x 9
* После 3 свёрточных слоёв: 13 x 13

Нужно очень много слоёв, если изображение размера 512 x 512.

Если мы берем фильтр большего размера, тогда поле восприятия тоже возрастает быстрее, но недостатчно быстро. Поэтому увеличение количества сверточных слоев и размера фильтра является недостаточным, выходной пиксель будет зависить от небольшого количества входных пикселей. Значит, нужно придумать что-то ещё, что позволит быстрее увеличивать размер поля восприятия. 

<h2 style="color:#008B8B">1.3 Свёртки с пропусками (strides)</h2>

### strides = 2

Параметр s отвечат за количество пропусков, если $s=1$, тогда у нас нет пропусков и свертки будут работать аналогично примерам выше.

<img src='img/lecture03/4.png'>

**Шаг 1.** Берем наш фильтор $3 \times 3$, после помещаем его с центром в левый верхний крайний угол входного изображения (первый пиксель). После мы сварачиваем область вокруг входного пикселя. 

**Шаг 2.** Если бы $s=1$, тогда цент фильтра сместился на один входной пиксель вправо, но так как $s=2$, центр фильтра будет смешён на два входных пикслея в право. Можем наблюдать это на картинке.

**Замечание:** пока мы не умеем сварачивать угловые входные пиксели. Рассмотрим это позже, представим что мы это умеем.

При использовании пропусков мы получаем картинку меньшего размера, но она тоже зависит от всей входной картинки (здесь речь о том, что мы для создания новой картинки используем все входные пиксели), так как для свернутого выходного пикселя имеется некоторая область восприятия, которая задействует пиксели при свертке. Другими словами, мы вокруг центра фильтра используем входные пиксели при свертке.

Но каждый выходной пиксель зависит от некоторой области восприятия, а не от всей картинки сразу.

### strides = 2 и сверткой $3 \times 3$

Применив к входному изображению свортку с $s=2$ и размером фильтра $3 \times 3$  и получим выходное изображение. К выходному изображению мы применяем свертку с размером фильтра $3 \times 3$ без использования пропусков.

<img src='img/lecture03/5.png'>

Какое поле восприятия у красного выходного пикселя? Тогда выходной красный пиксель зависит от входной синей области размером $3 \times 3$. И каждый пиксель синей обласит является выходным для первого сверточного слоя. Остается определить то, от какой входной области каждый выходной пиксель зависит и в совокупности мы получаем серую область, она получась более большой, так как мы использовали пропуски. А серая область определяет поле восприятия у красного выходного пикселя.

Добавление пропусков позволяет быстрее увеличивать поле восприятия для одного выходного пикселя.

Подробности про подсчёт размера поля: https://distill.pub/2019/computing-receptive-fields/

<h2 style="color:#008B8B">1.4 Dilated convolutions («раздутые» свёртки)</h2>

Если нам необходимо посчитать свертку первого выходного пикселя при использовании $l=2$ и фильтром размером $3 \times 3$. Тогда мы будем считать классическую свертку, но только с зелеными квадартами, которые мы выбрали на левом изображении. Следовательно в входном изображении мы будем пропускать некоторые пиксели входного изображения.
