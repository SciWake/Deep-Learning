<h1 style="color:black" align="center">Свёрточные сети</h1>

<h1 style="color:#008B8B">1 Receptive field</h1>

Имеется сверточный слой, который на вход получает картинку и на выход выдает картинку. Берем некоторый пиксель на выходной картинке и задаемся вопросм, от скольки пикселей входной картинки зависит данный выходной пиксель.

* Возьмём пиксель в итоговом изображении (после свёрточных слоёв)
* От какой части входного изображения зависит значение в этом пикселе?

<h2 style="color:#008B8B">1.2 Поле восприятия для свертки $3 \times 3$</h2>

### Один сверточный слой

Входную картинку мы сварачиваем с использованием фильтра 3 на 3 и получаем некоторую выходную картинку. Берем некоторый выходной пиксель и смотрим, от каких пикселей входного изображения он зависит. Чтобы получить данный пиксель, мы с центром входного пикселя (красный пиксель) поставили фильтр (зеленая область), после, этот входной пиксель с текущим фильтром мы свернули в выходной. Значит, у этого пикселя поле восприятия $3 \times 3$.

<img src='img/lecture03/1.png'>

### Два сверточных слоя

Левую картинку мы сварачиваем с использованием фильтра размером $3 \times 3$ и получаем картинку, которую мы сварачиваем с использованием другого фильтра размером $3 \times 3$.

<img src='img/lecture03/2.png'>

Какое поле восприятия у выходного красного пикселя? Красный пиксель мы получили на области из 9 пикселей промежуточного изображения (свернутого первым фильтром). 

А как мы получили каждый из этих 9 пикселей? Мы можем взять первый выходной пиксель, который обозначен фиолетовым цветом и посмотреть от каких входных пикселей он зависит: для этого берем входной пиксель с этими же координатами, относительно его центра на входном изображении надо поставить фильтр и свернем его в выходной фиоллетовый пиксель. Дальше берем каждый выходной пиксель первого сверточного слоя (зеленая область) и смотрим, от каких пикселей входного изображения он зависит и в совокупности мы получим синюю область.

Делаем выводы, что если применять несколько сверток подряд, тогда в каждом пикселе больше информации агрегируется о входном изображении.

**Поле восприятия для свёртки 3 x 3:**

* После 1 свёрточного слоя: 3 x 3
* После 2 свёрточных слоев: 5 x 5
* После 3 свёрточных слоёв: 7 x 7

Мы можем увеличивать размер поля восприятия за счет увеличения количества сверточных слоев подряд. Но размер поля восприятия возврастает довольно медленно с увеличением количества сверточных слоев подряд. Для этого можно вязть свертки большего размера:

<h2 style="color:#008B8B">1.2 Поле восприятия для свертки $5 \times 5$</h2>

### Два сверточных слоя

Левую картинку мы сварачиваем с использованием фильтра размером $5 \times 5$ и полученную картинку мы сварачиваем с использованием другого фильтра размером $5 \times 5$.

<img src='img/lecture03/3.png'>

Тогда выходной красный пиксель зависит от входной зеленой области размером $5 \times 5$. И каждый пиксель зеленой обласит является выходным для первого сверточного слоя. Остается определить то, от какой входной области каждый выходной пиксель зависит и в совокупности мы получаем синюю область. А синяя область определяет зону восприятия карсного пикселя выходного второго сверточного слоя.

**Поле восприятия для свёртки 5 x 5:**

Поле восприятия для свёртки 5 x 5:
* После 1 свёрточного слоя: 5 x 5
* После 2 свёрточных слоев: 9 x 9
* После 3 свёрточных слоёв: 13 x 13

Нужно очень много слоёв, если изображение размера 512 x 512.

Если мы берем фильтр большего размера, тогда поле восприятия тоже возрастает быстрее, но недостатчно быстро. Поэтому увеличение количества сверточных слоев и размера фильтра является недостаточным, выходной пиксель будет зависить от небольшого количества входных пикселей. Значит, нужно придумать что-то ещё, что позволит быстрее увеличивать размер поля восприятия. 

<h2 style="color:#008B8B">1.3 Свёртки с пропусками (strides)</h2>

### strides = 2

Параметр s отвечат за количество пропусков, если $s=1$, тогда у нас нет пропусков и свертки будут работать аналогично примерам выше.

<img src='img/lecture03/4.png'>

**Шаг 1.** Берем наш фильтор $3 \times 3$, после помещаем его с центром в левый верхний крайний угол входного изображения (первый пиксель). После мы сварачиваем область вокруг входного пикселя. 

**Шаг 2.** Если бы $s=1$, тогда цент фильтра сместился на один входной пиксель вправо, но так как $s=2$, центр фильтра будет смешён на два входных пикслея в право. Можем наблюдать это на картинке.

**Замечание:** пока мы не умеем сварачивать угловые входные пиксели. Рассмотрим это позже, представим что мы это умеем.

При использовании пропусков мы получаем картинку меньшего размера, но она тоже зависит от всей входной картинки (здесь речь о том, что мы для создания новой картинки используем все входные пиксели), так как для свернутого выходного пикселя имеется некоторая область восприятия, которая задействует пиксели при свертке. Другими словами, мы вокруг центра фильтра используем входные пиксели при свертке.

Но каждый выходной пиксель зависит от некоторой области восприятия, а не от всей картинки сразу.

### strides = 2 и сверткой $3 \times 3$

Применив к входному изображению свортку с $s=2$ и размером фильтра $3 \times 3$  и получим выходное изображение. К выходному изображению мы применяем свертку с размером фильтра $3 \times 3$ без использования пропусков.

<img src='img/lecture03/5.png'>

Какое поле восприятия у красного выходного пикселя? Тогда выходной красный пиксель зависит от входной синей области размером $3 \times 3$. И каждый пиксель синей обласит является выходным для первого сверточного слоя. Остается определить то, от какой входной области каждый выходной пиксель зависит и в совокупности мы получаем серую область, она получась более большой, так как мы использовали пропуски. А серая область определяет поле восприятия у красного выходного пикселя.

Добавление пропусков позволяет быстрее увеличивать поле восприятия для одного выходного пикселя.

Подробности про подсчёт размера поля: https://distill.pub/2019/computing-receptive-fields/

Следовательно в входном изображении мы будем пропускать некоторые пиксели входного изображения.

<h2 style="color:#008B8B">1.5 Pooling</h2>

Ещё один способ, который позволяет увеличивать поле восприятия.

Имеется входная картинка и давайте разобьём ее на блоки размером $2 \times 2$, получается у нас 9 блоков. И в каждом блоке мы возьмем максимум и запишем в выходную картинку вместо блока одно число:

<img src='img/lecture03/7.png'>

Данная операция называется Max-pooling с фильтром $2 \times 2$. Какой смысл? Если у нас имеется некоторый паттерн на картинке, тогда мы можем свернуть картинку с текущим фильтром и взять максимум. После смотрим на этот максимум и если он большой, тогда скорее всего на картинке находится данный паттерн.

Pooling позволяет сильно уменьшить картинку, не потеряв важную инфомрацию.

**Pooling + subsampling**

Pooling это более простая операция. При операции Pooling мы бы получили выходное изображении такого же размера, только бы в каждом блоке мы заменяли элемент на максиальный в блоке. Но обычно под Pooling ещё и subsampling, который мы обсудили выше.

**Замечание:** Обычно Pooling применяется уже к результату свертки, то есть, картинка слева это уже результат того, что мы входную картинку свернули с некоторым фильтром. Тогда, если у свертки был входной фильтр размера $3 \times 3$ и он детектировал диагонали, тогда мы смотрим, насколько в нашей области фильтр окликнулся на диагональный паттерн и записываем максимальное число отклика в новую ячейку.

### Выводы:

* Разбивает изображение на участки $n \times m$ и считает некоторую статистику в каждом участке (обычно максимум)
*  Существенно сокращает размер изображения (значит, увеличивает поле восприятия следующих слоёв)
* Не имеет параметров

**Зачем эти методы нужны?**

* Важно следить за тем, чтобы последние свёрточные слои имели размер поля восприятия, сравнимый со всей картинкой

<h1 style="color:#008B8B">2 Padding</h1>

* Если применять свёртку по формуле, то выходное изображение будет меньше входного

Входное изображение имеет размер $6 \times 6$, если размер фильтра будет $3 \times 3$, тогда после применения свертки мы получим выходное изображение размером $4 \times 4$. Так как мы не можем применять фильтр для крайних пикслей входного изображения.

<img src='img/lecture03/8.png'>


### Свертка в режиме Valid mode

* При честном подсчёте свёрток пиксели на краях не оказывают большого влияния на результат

<img src='img/lecture03/9.png'>

В процессе свертки не будет учитываться происходящее на краях картинки. Например, мы ищем паттерн который находится справа от входного изображения (уголок из еденицы справа и снизу). И мы видим, что у входного изображения слева в верхнем углу расположен кусочек данного паттерна. При обычной свертке мы не сможем поставить центр фильтра в карйний левый угол, где 0, значит мы не сможем обнаружить кусочек данного паттерна. Следовательно мы будем игнорировать некоторые паттерны на карях картинки.

<h2 style="color:#008B8B">1.1 Zero padding</h2>

Мы можем по карям картинки добавить нули, а после применяем свертку в режиме Valid mode.

<img src='img/lecture03/10.png'>

### Проблема:

* Добавляем по границам нули так, чтобы посчитанная после этого свёртка в valid mode давала изображение такого же размера, как исходное
* Есть риск, что модель научится понимать, где на изображении края — можем потерять инвариантность. При обучении нейронной сети, для нее нули могут стать хорошим индикатором того, что она находится с карю изображения, следовательно нейронная сеть может начать выучивать фильтры, которые специфичны только на карях картинки. А мы хотели чтобы сверточный слой был инвариантность к тому, где находится наш паттерн, чтобы для нас было не важно где он расположен, по середине, скраю... Следовательно, может быть такой фильтр, который будет регаировать на паттеры только границе, а мы хотим чтобы свертки были инвариантными к тому, где находится паттерн, а тут мы будем настраиваться на края.

<h2 style="color:#008B8B">1.3 Replication padding</h2>

В этом методе мы продалжаем заполнять клетки последним пикселем картинки.

<img src='img/lecture03/12.png'>

### Проблема :

* Пиксель на границе равен ближайшему пикселю из изображения

### Вывод:

* Паддинг позволяет контролировать размер выходных изображений
* Паддинг позволяет учитывать даже объекты на краях
* Разные типа паддингов допускают разные способы переобучения под края

<h2 style="color:#008B8B">1. Типичная архитектура</h2>

**1.** На вход поступает некоторая картинка, в нашем случае это черно белая картинка с одним каналом цвета $28 \times 28 \times 1$.

**2.** К входной картинке мы применяем сверточный слой без использования padding с размером фильтра $5 \times 5$ и на выходе будет изображение размером $24 \times 24 \times 1$. После свертки обычно используется нелинейность, которая позволяет усилить закономерности.

**3.** После применения свертки мы выполняем Max-pooling разбивая изображение на блоки размером $2 \times 2$ и на выходе будет изображение размером $12 \times 12 \times 1$. Который уменьшает размер картинки, тем самым увеличивая поле восприятия.

**4.** После применяем ещё один сверточный слой без использования padding с размером фильтра $5 \times 5$ и на выходе будет изображение размером $8 \times 8 \times 2$. Здесь мы используем два разынх фильтра, поэтому на выходе имее два канала.

**5.** После применения свертки мы выполняем Max-pooling разбивая изображение на блоки размером $2 \times 2$ и на выходе будет изображение размером $4 \times 4 \times 2$

<img src='img/lecture03/13.jpeg'>

**6.** На этом мы закончили процес свертки. После этого мы применяем операцию вытягивания (Flattened). Мы берем трехмерный массив и вытягиваем его в одномерный массив, получаем одномерный вектор. 

**7.** К этому одномерному вектору мы применяем два полносвязных слоя. На выходе последнего полносвязного слоя будет 10 чисел и мы требуем чтобы эти 10 чисел были вероятностями того, что соответствующее число находится на входной картинке.

Обычно после применения свертки добавляется некторая нелинейность, которые мы изучили и между полносвязными слоями тоже находится некоторая нелинейность.

Идея заключается в том, что мы сначала выполняем свертки, которые позволяют найти паттерны. Когда мы считаем что количество паттернов достаточное, тогда всю эту информацию мы можем совместить с помощью полносвязных слоев и решить задачу.

https://towardsdatascience.com/a-comprehensive-guide-to-convolutional-neural-networks-the-eli5-way-3bd2b1164a53

### Типичная архитектура

* Последовательное применение комбинаций вида «свёрточный слой -> нелинейность -> pooling» или «свёрточный слой -> нелинейность»
* Выпрямление (flattening) выхода очередного слоя
* Серия полносвязных слоёв
